```python
# æ–¹æ³•1 æ”¹è¿›ç‰ˆ
count = 0
class Person:
    def __init__(self):
        # åªèƒ½è¯»å–å…¨å±€å˜é‡ï¼Œä¸èƒ½ä¿®æ”¹ï¼Œå› ä¸ºä¿®æ”¹ä¼šäº§ç”Ÿæ­§ä¹‰:
        # åˆ°åº•æ˜¯å†…éƒ¨æ–°å¢äº†ä¸€ä¸ªå˜é‡ï¼Œè¿˜æ˜¯ç”¨çš„å¤–é¢çš„é‚£ä¸ª
        global count
        print("è®¡æ•° + 1")
        count += 1

    def __del__(self):
        global count
        print("è®¡æ•° - 1")
        count -= 1
        
    def log(self):  # æŠŠprintè¯­å¥æ”¶è¿›logæ–¹æ³•ä¸­
        print("å½“å‰äººçš„%dä¸ª"%count)

p = Person()
p2 = Person()
p.log()
del p2
p2.log()  # p2è°ƒç”¨å¤±è´¥
```

```python
# æ–¹æ³•1 æ”¹è¿›ç‰ˆ2
count = 0
class Person:
    def __init__(self):
        # åªèƒ½è¯»å–å…¨å±€å˜é‡ï¼Œä¸èƒ½ä¿®æ”¹ï¼Œå› ä¸ºä¿®æ”¹ä¼šäº§ç”Ÿæ­§ä¹‰:
        # åˆ°åº•æ˜¯å†…éƒ¨æ–°å¢äº†ä¸€ä¸ªå˜é‡ï¼Œè¿˜æ˜¯ç”¨çš„å¤–é¢çš„é‚£ä¸ª
        global count
        print("è®¡æ•° + 1")
        count += 1

    def __del__(self):
        global count
        print("è®¡æ•° - 1")
        count -= 1
        
    @staticmethod    # æ²¡ç”¨è¿‡self, æ”¹æˆé™æ€æ–¹æ³•
    def log():  
        print("å½“å‰äººçš„%dä¸ª"%count)

p = Person()
p2 = Person()
Person.log()  # æ³¨æ„è¿™é‡Œï¼Œç”¨Personç±»è°ƒç”¨ã€‚åœ¨pycharmä¸Šæœ‰æ—¶æ˜¯2
del p
Person.log()
```

ğŸ‘†å…¨å±€å˜é‡æ„å‘³ç€åœ¨å“ªéƒ½èƒ½æ“ä½œæ”¹å˜ä»–ï¼Œä¸å®‰å…¨ï¼Œè¿èƒŒäº†ç±»çš„å°è£…æ€§

```python
class Person:
    
    __count = 0     # ç±»çš„ç§æœ‰å˜é‡
    
    def __init__(self):
        print("è®¡æ•° + 1")
        Person.__count += 1

    def __del__(self):
        print("è®¡æ•° - 1")
        self.__class__.__count -= 1
        
    @classmethod    # æ”¹æˆç±»æ–¹æ³•
    def log(cls):  
        print("å½“å‰äººçš„%dä¸ª" % cls.__count)

p = Person()
p2 = Person()
Person.log()  # æ³¨æ„è¿™é‡Œï¼Œç”¨Personç±»è°ƒç”¨ï¼Œåœ¨pycharmä¸Šæœ‰æ—¶æ˜¯2
del p
Person.log()
```
